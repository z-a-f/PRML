# 1.1 Polynomial Curve Fitting

```python
sigma = 0.2
step = 0.01
start = 0.
stop = 1.
x = np.arange(start, stop+step, step)
t_true = np.sin(2*np.pi*x)
t_noisy = t_true + np.random.normal(0, sigma, t_true.size)
plt.plot(x, t_true, label='$t = \sin(2\pi x)$')
plt.plot(x, t_noisy, label='$t = \sin(2\pi x)+\mathcal{N}(0, %.1f)$'%sigma)
plt.legend()
plt.xticks([0,1])
plt.yticks([-1,0,1])
plt.xlabel('$x$')
plt.ylabel('$t$')
```

```python
x_15 = np.random.permutation(x.size)[:15]
x_50 = np.random.permutation(x.size)[:50]
x_100 = np.random.permutation(x.size)[:100]

fig, ax = plt.subplots(1,3, figsize = [12,4], sharey=True)

ax[0].plot(x, t_true)
ax[0].plot(x[x_15], t_noisy[x_15], 'o', ms=8)
ax[0].set_title('15 samples')

ax[1].plot(x, t_true)
ax[1].plot(x[x_50], t_noisy[x_50], 'o', ms=8)
ax[1].set_title('50 samples')

ax[2].plot(x, t_true)
ax[2].plot(x[x_100], t_noisy[x_100], 'o', ms=8)
ax[2].set_title('100 samples')

for idx in xrange(3):
  ax[idx].set_xlim([-0.1,1.1])
  ax[idx].set_ylim([-1.2,1.2])
  ax[idx].set_xticks([0,1])
  ax[idx].set_yticks([-1,0,1])
ax[1].set_xlabel('$x$')
ax[0].set_ylabel('$t$')
```
